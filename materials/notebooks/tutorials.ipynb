```python
import pandas

df = pandas.read_csv("./Library_Usage.csv")
```

# Tutorial - Datensätze filtern

Letzte Woche haben wir gesehen, wie einzelne Spalten eines Datensatzes ausgewählt und einer neuen Variablen zugeordnet werden können:

```python
checkouts = df['Total Checkouts']
```

Um nur Beobachtungen mit mehr als *10000* Ausleihen zu betrachten, können wir den Datensatz filtern:

```python
many_checkouts = checkouts > 10000
df[many_checkouts]
```

> - `many_checkouts` enthält ein Objekt vom Typ `pandas.Series`. Die Werte sind `True`, wenn der entsprechende Wert in `checkouts` größer *500* ist. Die Werte sind `False`, wenn der entsprechende Wert in `checkouts` kleinergleich *500* ist. Diese Art von `pandas.Series` nennt man auch `boolean array` (logische Liste/ Reihe). 
- Setze man einen `boolean array` ein, um einen `dataframe` zu filtern, dann werden nur die Zeilen zurückgegeben, bei denen der array `True` ist. 

Die Befehle lassen sich auch ohne Hilfe der Variablen schreiben:

```python
df[checkouts > 10000]
# or
df[df['Total Checkouts'] > 10000]
```

Der gefilterte Datensatz lässt sich natürlich auch wieder einer neuen Variablen zuordnen. Damit kann dann wieder gerechnet werden!

```python
df_filter = df[checkouts > 10000]
df_filter['Age Range'].value_counts(normalize=False)
```

# Tutorial - Datensätze grupieren 

Mit der `group_by` Funktion können Datensätze nach den Werten in einer bestimmten Spalte gruppiert werden. Der gruppierte Datensatz kann dann in einer Variablen zwischengespeichert werden. Mit ihm lässt sich rechnen wie bisher:

```python
df_grouped = df.groupby('Patron Type Definition')
checkouts = df_grouped['Total Checkouts']
checkouts.sum()
```

```python
checkouts.mean().round()
```
