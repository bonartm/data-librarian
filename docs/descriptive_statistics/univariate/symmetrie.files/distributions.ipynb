# Visualization of distributional shapes using the [beta-distribution](https://en.wikipedia.org/wiki/Beta_distribution)

```python
import numpy as np
import seaborn as sns
import pandas as pd
%matplotlib inline
sns.set()

dat = pd.DataFrame()
vals = [0.8, 1, 3]
for alpha in vals:
    for beta in vals:
        x = {
            'x': np.random.beta(alpha+0.1, beta+0.1, size = 10000),
            'alpha': alpha,
            'beta': beta
        }
        dat = dat.append(pd.DataFrame(x))

g = sns.FacetGrid(dat, col="alpha", row="beta", sharey=False, height=8)
g = g.map(sns.distplot, "x")
g.savefig("./distributions.png", dpi=150)
g
```
